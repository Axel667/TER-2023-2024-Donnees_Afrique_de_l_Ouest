```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import pydeck as pdk
import numpy as np
```

```python
citations_pays = pd.read_csv("/kaggle/input/data-geo-pop/countries_mentions.csv")
citations_pays.info()
citations_pays
```

```python
citations_pays['Pays'] = citations_pays['Pays'].replace({'Cap-Vert': 'Cabo Verde'})
citations_pays
```

```python
wab = gpd.read_file("/kaggle/input/data-geo-pop/world-administrative-boundaries.geojson")
wab
```

```python
wab = wab.rename(columns={'french_short': 'Pays'})
wab
```

```python
geo_citations_pays = wab.merge(citations_pays, how="left")
geo_citations_pays
```

```python
geo_citations_pays.to_file("west_af_citations.json", driver="GeoJSON")
```

```python
west_af_citations = gpd.read_file("west_af_citations.json")
west_af_citations
```

```python
west_af_citations.info()
```

```python
west_af_citations["citations_log"] = np.log(citations_pays['citations'] + 0.01)
west_af_citations['citations_log_norm'] = ((west_af_citations['citations_log'] - west_af_citations['citations_log'].min()) /
                                 (west_af_citations['citations_log'].max() - west_af_citations['citations_log'].min()))
west_af_citations
```

```python
Vue = pdk.ViewState(latitude=12.2779298, 
                                   longitude=-1.74014113189,
                                   zoom=5, max_zoom=16, pitch='100', bearing=0)

carte = pdk.Layer(
    "GeoJsonLayer",
    west_af_citations,
    opacity=1,
    stroked=True,
    filled=True,
    extruded=True,
    wireframe=True,
    pickable=True,
    get_elevation= "citations",
    get_fill_color="[255 * citations_log_norm, 255 * (1 - citations_log_norm), 0]",
    get_line_color="[50, 50, 50]"
)

r = pdk.Deck(layers=[carte], initial_view_state=Vue)

r.to_html("map.html")
```
